```python
nums = [2, 7, 11, 15]
target = 9

def two_sum(nums, target):
    seen = {}
    for i, x in enumerate(nums):
        if target - x in seen:
            return [seen[target-x], i]
        seen[x] = i # add the key value pair into the seen dictionary
    return []
print(two_sum(nums, target))

# The key is the number we have enountered, the value is the index
# So, seen[target - x] is saying: "Tell me the index of the number that, when added to my current number, equals the target."
# We are using the complement which is the key tp find the index of the complement which is the value
```

```python
#We want to group anagrams, that is word that contain the same letters
#Create a function that takes in a list of strings and output a list of list, with each inner list containing the anagrams

from collections import defaultdict

def group_anagrams(words):
    groups = defaultdict(list)
    # sorted(word) will return a list of seperate elements in sorted order
    #"".join() joins elemnts in a list with the empty string as the glue
    for w in words:
        key = "".join(sorted(w)) #the key is the sorted string
        groups[key].append(w) #append the word w into the list that is associated with the key, the sorted string
                              #this way words with the same sorted string are in the same list, belong to the same key
    return list(groups.values()) #return the values in a list

print(group_anagrams(["eat","tea","tan","ate","nat","bat"]))
    
    
```

### Contiguous subarrays means that the elements in the subarray must be adjecent in the original list

```python
def subarray_sum(nums, k):
    count = 0
    prefix = 0
    seen = {0:1} #the key represent the prefix sum, the value represents how many times it has appeard
    for x in nums:
        prefix += x
        count += seen.get(prefix-k, 0) #currently the capacity is prefix. We want to see if there is another previous prefix that equal current prefix - k. this is because the difference between them is k
        seen[prefix] = seen.get(prefix, 0)+1 #we check how many times the prefix sum has appeard and add 1 to it for the current
    return count
print(subarray_sum([1, 1, -1, 2, -1, 2], 2))
```

```python
#Longest consecutive sequence
# turn the unsorted list into a set
unsorted = [100, 2, 4, 3, 1]
print(sorted(set(unsorted)))

def longest_consecutives(nums):
    s = sorted(set(nums))
    best = 0
    for x in s:
        if x-1 not in s: #it is the first in the sequence
            length = 1
            while x + length in s: #continue the check be extending the sequence
                length += 1
            best = max(best, length) #comparre to other sequence lengths
    return best

print(longest_consecutives(unsorted))
```

```python
# We want to find the indeces of the elements in a sorted list that equal the target
def two_sum_sorted(arr, target):
    i = 0
    j = len(arr) - 1 #to get the index of the last element
    while i < j: #while the pointers have not crossed each other
        s = arr[i] + arr[j]
        if s == target:
            return (i,j)
        elif s < target: #we need to use a bigger smallest number
            i += 1
        else: #only case is if s is greater than the target
            j -= 1 # reduce the biggest element
    return None #will reach this if there are no combination
print(two_sum_sorted([1, 3, 5, 9, 7], 10))

# We do no need a dictionary as the list is sorted. 

```

```python
# For three sum, we fix one and use pointer for the two others. 
# this require us to sort the list. 
def three_sum(nums, target):
    nums.sort()
    results = []
    n = len(nums)
    for i in range(n-2): #we exclude the two pointer start position
        if i > 0 and nums[i] == nums[i-1]:
            continue #i is the anchor, when choosing the next anchor we do not check the original anchor
        # this is because we are not interested in the index of the numbers but the values themselves. So if the number is the same for tow anchors then we skip. We can check by looking at the value next to it since the list is sorted
        j = i+1 #assign the pointers. the anchor is the smallest and the pointer look for solutions for values that are bigger
        k = n-1
        while j<k: # while the pointer can still check new combinations
            total = nums[i] + nums[j] + nums[k]
            if total == target:
                results.append([nums[i], nums[j],nums[k]])
                j += 1 #we need to move both pointers since moving only one would result in a sum that is smaller or larger than the current which is equal to the target
                k-=1
                while j < k and nums[j] == nums[j-1]:
                    j += 1 #if the moved number is the same as the original we need to move again since we are not interestd in the same number
                while j < k and nums[k] == nums[k+1]:
                    k -= 1
            elif total < target: #if the is not equal in the first place
                j+=1
            else:
                k -= 1
    return results

print(three_sum([-1, 0, 1, 2, -1, -4], 0))
```

```python
def max_area(heights):
    i = 0
    j = len(heights)-1
    best = 0 # the biggest area
    while i < j:
        area = min(heights[i], heights[j])*(j-1)
        best = max(best,area)
        if heights[i] < heights[j]: #when the pointers have not crossed
            i+=1
        else:
            j -= 1
    return best
#the goal is the move the shorter line inwards 
print(max_area([1, 8, 6, 2, 5, 4, 8, 3, 7]))
```

```python
from functools import cache


@cache
def fibonachi(n): #find the n:th fibonachi number
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonachi(n-1) + fibonachi(n-2)

print(fibonachi(1000))
```
